```python
import spacy

# Load the model
nlp = spacy.load("en_core_web_sm")

# Process text
doc = nlp("Apple is buying U.K. startup for $1 billion.")

# Print detected entities
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python

import spacy

# Load the model
nlp = spacy.load("en_core_web_sm")

# Process text (Fixed the line break at the end of the string)
doc = nlp(
    "Dr. Jane Ndagire, the Chief Medical Officer at Kampala City Hospital, "
    "announced on Friday that the institution has partnered with the Uganda "
    "Ministry of Health to roll out 500 new mobile health clinics. The initiative, "
    "funded by a $12 million grant from the World Health Organization, aims to "
    "improve healthcare delivery across the East African region. The first fleet "
    "of trucks will be manufactured by Toyota East Africa in Nairobi and shipped "
    "to Entebbe by October."
)

# Print detected entities
for ent in doc.ents:
    print(f"{ent.text:<30} {ent.label_}")
```
